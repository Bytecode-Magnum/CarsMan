#### Car Manufactured DataAnalysis

#### Importing the necessary libraries for the analysis

```python
import numpy as np
import pandas as pd
df=pd.read_csv('./data/car_data.csv')
```

```python
df.head()
```

#### Columns and their data types

```python
df.info()
#  as we can see all tha columns have appropriate data types , so no need to change that
```

#### changing column names

```python
df.rename(columns={'Engine Fuel Type':'Engine_Fuel_Type', 'Engine HP':'Engine_HP',
       'Engine Cylinders':"Engine_Cylinders", 'Transmission Type':"Transmission_Type",
       'Number of Doors':'Doors', 'Market Category':'Market_Category', 'Vehicle Size':"Vehicle_Size", 'Vehicle Style':'Vehicle_Style',
       'highway MPG':'Highway_mpg', 'city mpg':'City_mpg'},inplace=True)
```

#### Exploring the null rows

```python
df.loc[df.isnull().any(axis=1)]
df.dropna(inplace=True)
```

#### Exploring the managing duplicate rows

```python
df.loc[df.duplicated(keep=False)]
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

#### Name of different makers 

```python
df.Make.unique()
```

#### from each makers display their all the model names

```python
makers=df.Make.unique()
for i,make in enumerate(makers):
    name=make
    models=df.query(f'Make=="{name}"')['Model'].unique()
    data={
        "Makers":name, 'Models':models
    }
    print((i+1),data)
```

#### Number of cars manufactured by various makers

```python
df.Make.value_counts()
# so most of the cars are manufactured by Chevrolet, followed by Fordm Volkswagen and Cadillac
```

```python
np.sort(df.Year.unique())
```

```python
df.head()
```

#### Who manufactured most cars between 2012 and 2017

```python
df.query("2012<Year<=2017").Make.value_counts()
```

#### Number of AUTOMATED_MANUAL AUTOMATIC DIRECT_DRIVE and MANUAL manufactured by different makers

```python
df=df[~(df.Transmission_Type=="UNKNOWN")]
df.groupby('Make').Transmission_Type.value_counts().unstack('Transmission_Type').fillna(0).sort_values(by='AUTOMATED_MANUAL',ascending=False).head()
# So Volkswagen most number of automated_manual cars
```

```python
df.groupby('Make').Transmission_Type.value_counts().unstack('Transmission_Type').fillna(0).sort_values(by='AUTOMATIC',ascending=False).head()
# and chevrolet manufacture manufacture maximum number of automatic cars
```

#### Number of all wheel drive	four wheel drive	front wheel drive	rear wheel drive cars manufactured by different makers

```python
df.groupby('Make').Driven_Wheels.value_counts().unstack('Driven_Wheels').fillna(0).sort_values(by='all wheel drive',ascending=False).fillna(0).head()
# so audi manufacture most number of all wheel driver cars
```

#### Most popular car over past 5 years

```python
df.query('2013<=Year<=2017').pivot_table(index='Make',columns="Year",values="Popularity",aggfunc=('max')).sort_values(by=2017,ascending=False).fillna(0)
# so over the past 5 years the popularity of ford cars were at the peak
```

#### Who makes the maximum number of large cars

```python
df.groupby('Make')['Vehicle_Size'].value_counts().unstack('Vehicle_Size').fillna(0).sort_values(by='Large',ascending=False).head()
```

#### Maker and their heighest engine hp cars

```python
df.groupby(['Make','Model'])['Engine_HP'].agg(HP=('max')).sort_values(by='HP',ascending=False).head()
# Bugatti Veyron has heighest hp followed by lamborghini
```

#### From past 3 years maker with heighest hp and highway mpg

```python
i=df.query('2015<=Year<=2017').pivot_table(index=["Make"],columns='Year',values=['Engine_HP','Highway_mpg'],aggfunc="max").fillna(0).head()
i.sort_values(by=[('Engine_HP',2017),('Highway_mpg',2017)],ascending=False)
```

#### finding the corelation between engine_hp engine_Cylinder , highway_mpg, city_mpg and popularity

```python
sub=df[['Engine_HP','Highway_mpg','Popularity','City_mpg','Engine_Cylinders']]
sub.corr()
```

#### Who manufactured maximum number of Coupe style cars over the years

```python
df.query('Vehicle_Style=="Coupe"').groupby('Make')['Make'].agg(Total=('count')).nlargest(5,columns='Total')
```

#### Number of cars manufactured by different makers past 3 years

```python
df.query('2015<=Year<=2017').groupby('Year').Make.value_counts().unstack('Year').fillna(0)
```
